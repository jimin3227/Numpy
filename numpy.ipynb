## 배열 속성 정보

```python
def array_info(array):
    print(array)
    print("ndim:",array.ndim)
    print("shape:",array.shape)
    print("dtype:",array.dtype)
    print("size:",array.size)
    print("itemsize(byte):",array.itemsize)
    print("nbytes(size*itemsize):",array.nbytes)
    print("strides:",array.strides)
```

# 배열 값 삽입/수정/삭제/복사

## 배열 값 삽입

- insert():배열의 특정 위치에 값 삽입
- 원본 배열 변경없이 새로운 배열 반환
- axis를 지정하지 않으면 1차원 배열로 변환 후 삽입

```python
print(a1)
b1 = np.insert(a1,0,10)
print(b1)
```

```python
print(a2)
b2 = np.insert(a2,1,10)
print(b2)
c2 = np.insert(a2,1,10,axis=0)
print(c2)
d2 = np.insert(a2,1,10,axis=1)
print(d2)
```

## 배열 값 수정

```python
print(a1)
a1[0]=1
a1[1]=2
a1[2]=3
print(a1)
a1[:2]=9 # 슬라이싱
print(a1)
idx = np.array([0,2]) # 팬시 인덱싱
a1[idx]=0
print(a1)
a1[idx]+=4
print(a1)
```

```python
print(a2)
a2[0,0]=1
a2[1,1]=2
a2[2,2]=3
print(a2)
a2[0]=1
print(a2)
a2[1:,2]=9 # 슬라이싱
print(a2)
row = np.array([0,1]) # 팬시 인덱싱
col = np.array([1,2])
a2[row,col]=0
print(a2)
```

## 배열 값 삭제

- delete():배열의 특정 위치에 값 삭제
- 원본 배열 변경없이 새로운 배열 반환
- axis를 지정하지 않으면 1차원 배열로 변환 후 삭제

```python
print(a1)
b1 = np.delete(a1,1)
print(b1)
```

```python
print(a2)
b2 = np.delete(a2,1)
print(b2)
c2 = np.delete(a2,1,axis=0)
print(c2)
d2 = np.delete(a2,1,axis=1)
print(d2)
```

## 배열 복사

- 리스트와 달리 슬라이싱해도 `copy()`가 아닌 이상 원본 배열에 영향

```python
print(a2)
a2_sub=a2[:2,:2]
print(a2_sub)
a2_sub[:,1]=0
print(a2_sub)
print(a2)
```

```python
print(a2)
a2_sub_copy=a2[:2,:2].copy()
print(a2_sub_copy)
a2_sub_copy[:,1]=1
print(a2_sub_copy)
print(a2)
```

## 배열 분할

- `split()`: 배열 분할

```python
a1=np.arange(0,10)
print(a1)
b1,c1=np.split(a1,[5])
print(b1,c1)
b1,c1,d1,e1,f1=np.split(a1,[2,4,6,8])
print(b1,c1,d1,e1,f1)
```

- `vsplit()`: 수직 분할, 1차원으로 분할

```python
a2=np.arange(1,10).reshape(3,3)
print(a2)
b2,c2=np.vsplit(a2,[2])
print(b2)
print(c2)
```

- `hsplit()`: 수평 분할, 2차원으로 분할

```python
a2=np.arange(1,10).reshape(3,3)
print(a2)
b2,c2=np.hsplit(a2,[2])
print(b2)
print(c2)
```

- `dsplit()`: 깊이 분할, 3차원으로 분할

```python
a3=np.arange(1,28).reshape(3,3,3)
print(a3)
b3,c3=np.dsplit(a3,[2])
print(b3)
print(c3)
```

## 비교 연산(Comparison Operators)

```python
a1=np.arange(1,10)
print(a1)
print(a1==5)
print(a1!=5)
print(a1<5)
print(a1<=5)
print(a1>5)
print(a1>=5)
```

```python
a2=np.arange(1,10).reshape(3,3)
print(a2)
print(np.count_nonzero(a2>5))
print(np.sum(a2>5))
print(np.sum(a2>5,axis=0))
print(np.sum(a2>5,axis=1))
print(np.any(a2>5))
print(np.any(a2>5,axis=0))
print(np.any(a2>5,axis=1))
print(np.all(a2>5))
print(np.all(a2>5,axis=0))
print(np.all(a2>5,axis=1))
```

```python
a1=np.array([1,2,3,4,5])
print(a1)
b1=np.array([1,2,3,3,4])
print(b1)
print(np.isclose(a1,b1))
```

```python
a1=np.array([np.nan,2,np.inf,4,np.NINF])
print(a1)
print(np.isnan(a1))
print(np.isinf(a1))
print(np.isfinite(a1))
```

```python
a2=np.arange(1,10).reshape(3,3)
print(a2)
print((a2>5)&(a2<8))
print(a2[(a2>5)&(a2<8)])
print((a2>5)|(a2<8))
print(a2[(a2>5)|(a2<8)])
print((a2>5)^(a2<8))
print(a2[(a2>5)^(a2<8)])
print(~(a2>5))
print(a2[~(a2>5)])
```

## 배열 정렬

```python
a1=np.random.randint(1,10,size=10)
print(a1)
print(np.sort(a1)) # 원배열 변화 X
print(np.argsort(a1)) # sort된 것의 원래 index
print(a1.sort()) # 원배열 변화 O
```

```python
a2=np.random.randint(1,10,size=(3,3))
print(a2)
print(np.sort(a2,axis=0))
print(np.sort(a2,axis=1))
```

### 부분 정렬

- `partition()`: 배열에서 가장 작은 값 k개 반환

```python
a1=np.random.randint(1,10,size=10)
print(a1)
print(np.partition(a1,3))
```

```python
a2=np.random.randint(1,10,size=(4,4))
print(a2)
print(np.partition(a2,2)) # default:axis=1(가로방향)
print(np.partition(a2,2,axis=0))
print(np.partition(a2,2,axis=1))
```

# 배열 입출력

```python
a2=np.random.randint(1,10,size=(5,5))
print(a2)
np.save('a2',a2)
```

```python
npy=np.load('a2.npy')
print(npy)
```

```python
b2=np.random.randint(1,10,size=(5,5))
print(b2)
np.savez('a2b2',a2,b2)
```

```python
npz=np.load('a2b2.npz')
print(npz.files)
print(npz['arr_0'])
print(npz['arr_1'])
```
